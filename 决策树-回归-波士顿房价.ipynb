```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, cross_val_score
import pandas as pd
import numpy as np
```

```python
boston = pd.read_csv('house_data.csv') #波士顿房价模型已经从sklearn.datasets库中删除了，出于某些政治正确
boston
```

```python
regressor = DecisionTreeRegressor(random_state=0) #实例化
cross_val_score(regressor, boston.drop('MEDV', axis=1), boston['MEDV'], cv=10, 
                scoring='neg_mean_squared_error') #如果默认什么都不填，返回的是R^2，但是我们一般使用MSE

'''r方越接近1越好
mse显示为负数
mse越接近0越好'''
```
